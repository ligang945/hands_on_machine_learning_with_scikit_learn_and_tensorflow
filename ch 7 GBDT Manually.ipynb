```python
import numpy as np
import matplotlib.pyplot as plt

def lg_plot(yy):
    plt.figure()
    plt.plot(X, y, "b.")
    plt.plot(X, yy, "r")
    plt.axis([-3, 3, -3, 20])
    plt.show()
```

```python
# 生成数据，带噪声的二次曲线
m = 100
X = np.arange(-3, 3, 6/m).reshape((m, 1))
y = 2 * X**2 + np.random.randn(m, 1)
print(X.shape)
print(y.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg1 = DecisionTreeRegressor(max_depth=2)
tree_reg1.fit(X, y)
y1 = tree_reg1.predict(X).reshape((m, 1))
print(y1.shape)
lg_plot(y1)
```

```python
tree_reg2 = DecisionTreeRegressor(max_depth=2)
tree_reg2.fit(X, y - y1)
y2 = tree_reg2.predict(X).reshape((m, 1))
print(y2.shape)
lg_plot(y2)
```

```python
lg_plot(y1 + y2)
```

```python
tree_reg3 = DecisionTreeRegressor(max_depth=2)
tree_reg3.fit(X, y - y1 - y2)
y3 = tree_reg3.predict(X).reshape((m, 1))
print(y3.shape)
lg_plot(y3)
```

```python
lg_plot(y1 + y2 + y3)
```

```python
# GBDT
from sklearn.ensemble import GradientBoostingRegressor

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X, y.ravel())
y_gb = gbrt.predict(X).reshape((m, 1))
print(y_gb.shape)
lg_plot(y_gb)
```
